# Plot of the various versions of the damped oscillator
For section 8.2

Solutions:
$$ \begin{align}
 x_\mathrm{undamped}(t) &= x(0) \cos(\omega_0 t) + \frac{v(0)}{\omega_0} \sin(\omega_0 t), \\
 x_\mathrm{underdamped}(t) &= e^{-\zeta \omega_0 t} \left[A \cos(\omega_\mathrm{d} t) + B \sin(\omega_\mathrm{d} t) \right], \\
 x_\mathrm{critdamped}(t) &= (A + Bt) e^{-\omega_0 t}, \\
 x_\mathrm{overdamped}(t) &= e^{-\zeta \omega_0 t} \left[A e^{\Omega t} + B e^{-\Omega t}\right].
\end{align}
$$

```python
# Python libraries.

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def x(t):
    # Returns x(t), with x(0) = 1, v(0) = 0, and omega_0 = 1.
    return np.cos(t)

def xud(t):
    # Returns x_{underdamped}(t), zeta = 0.25 and omega_0 =1 (and therefore omega_d = sqrt{1-0.25^2}).
    return np.exp(-0.25*t) * np.cos(np.sqrt(1-0.25*0.25) * t)

def xcd(t):
    # Returns x_{critically damped}(t), A = B = omega_0 = 1.
    return (1 + t) * np.exp(-t)

def xod(t):
    # Returns x_{overdamped}(t), with zeta = 2 and omega_0 =1, and therefore Omega = sqrt(3). Also choose A and B such that x(0) = 1.
    return ((0.5 + 1/np.sqrt(3))*np.exp(np.sqrt(3) * t) + (0.5 - 1/np.sqrt(3))*np.exp(-np.sqrt(3) * t)) * np.exp(-2*t)
```

```python
fig, ax = plt.subplots(figsize=(6,4))

t = np.linspace(0, 4*np.pi, 1000)

line1 = ax.plot(t, x(t), label='$\\zeta = 0$')
line2 = ax.plot(t, xud(t), label='$0 < \\zeta < 1$')
line3 = ax.plot(t, xcd(t), label='$\\zeta = 1$')
line4 = ax.plot(t, xod(t), label='$\\zeta > 1$')

ax.axhline(y = 0, color = 'k', linestyle = ':')

ax.set_xlim(0,4*np.pi)
ax.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi], labels=['$0$', '$\\pi$', '$2\\pi$', '$3\\pi$', '$4\\pi$'])
ax.set_yticks([-1, -0.5, 0, 0.5, 1], labels=['$-1$', '$-\\frac{1}{2}$', '$0$', '$\\frac{1}{2}$', '$1$'])
#plt.ylim(-0.2,1.0)
ax.set_xlabel('$t$')
ax.set_ylabel('$x(t)$')
ax.legend()

#plt.savefig("dampedoscillations.pdf", format="pdf", bbox_inches="tight")
#plt.savefig("dampedoscillations.svg", format="svg", bbox_inches="tight")
```
